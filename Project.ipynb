```python
#extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract file to dir
zip_ref.close() #close file
```

```python
#read data from the file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv')
#df
df['Ship Mode'].unique()
```

```python
#handling Not Available and unknown values
df = pd.read_csv('orders.csv',na_values = ['Not Available', 'unknown'] )
df['Ship Mode'].unique()
```

```python
#rename column names and make them lower case and replace spce with underscore e.g. ship_mode
#df.rename(columns = {'Order Id':'order_id', 'City':'city'}) #not a good way to rename as user has to put in all the names
df.columns = df.columns.str.lower()
df.columns= df.columns.str.replace(' ', '_')
df.head(5)
```

```python
#derive new column (discount, sales price, and profit)
df['discount'] = df['list_price'] * df['discount_percent'] * 0.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
df.dtypes
```

```python
#convert order_date data type from object to datetime
df['order_date']=pd.to_datetime(df['order_date'], format = '%Y-%m-%d')
```

```python
#drop list price cost price and discount percent columns
df.drop(columns = ['cost_price', 'list_price', 'discount_percent'], inplace=True)
```
